Now we can perform preprocessing of the data – change to lower case, remove stopwords using NLTK, and remove numbers.

Tokenise the data using word_tokenize from NLTK.

```python
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk import sent_tokenize, word_tokenize
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
from nltk.tokenize import RegexpTokenizer

import seaborn as sns
import re
```

```python
#### our clean_text function will lower case the text, remove stopwords and numbers using nltk methods.
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub('[^A-Za-z]+', ' ', text)  # Remove numbers and punctuation
    tokens = word_tokenize(text)  # Tokenize words
    stop_words = set(stopwords.words('english'))

    # Remove stopwords
    filtered_tokens = [word for word in tokens if word not in stop_words]

    # Remove short words (≤2 characters)
    filtered_tokens = [word for word in filtered_tokens if len(word) > 2]

    return " ".join(filtered_tokens)  # Return cleaned text as a string
```

```python
###### let's create a new column in each data set called cleaned_cleaned review with the clean_text function
trustpilot_df_by_days["cleaned_review"] = trustpilot_df_by_days["Review Content"].apply(clean_text)
google_df_by_days["cleaned_review"] = google_df_by_days["Comment"].apply(clean_text)
```

Find the frequency distribution of the words from each data set's reviews separately. You can use nltk.freqDist.

Plot a histogram/bar plot showing the top 10 words from each data set.

```python
import nltk
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
##### Now, let's define a function that creates a bar plot with the frequency of the top words in the reviews.
nltk.download("punkt")

def plot_word_frequency(df, key, top_n=20, remove_words=None):
    """
    Plots the frequency distribution of the top words in the cleaned reviews.

    Parameters:
        df (pd.DataFrame): The dataframe containing cleaned reviews.
        key (str): The column name containing cleaned review text.
        top_n (int): Number of top words to display.
        remove_words (list): List of words to exclude from the frequency distribution.
    """
    # Tokenize and flatten all words
    all_words = " ".join(df[key]).split()

    # Remove specified words if provided
    if remove_words:
        all_words = [word for word in all_words if word.lower() not in remove_words]

    # Get word frequency distribution
    word_distribution = nltk.FreqDist(all_words)

    # Convert to DataFrame
    freq_df = pd.DataFrame(word_distribution.most_common(top_n), columns=["Word", "Frequency"])

    # Plot using seaborn
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Word", y="Frequency", data=freq_df, palette="viridis")
    plt.xticks(rotation=45)
    plt.title(f"Top {top_n} Most Frequent Words")
    plt.show()

# Apply function to dataset
plot_word_frequency(trustpilot_df_by_days, "cleaned_review", top_n=10)
```

```python
plot_word_frequency(google_df_by_days, "cleaned_review", top_n=10)
```

Use the wordcloud library on the cleaned data and plot the word cloud.


```python
def word_cloud_display(df, key, remove_words=None):
  #### convert reviews into a single string
  all_words = " ".join(df[key])
  ### generate word cloud
  # Remove specified words
  if remove_words:
        for word in remove_words:
            all_words = all_words.replace(f" {word} ", " ")
  wc = WordCloud(background_color = 'black', max_words = 1000,  max_font_size = 50)
  plt.figure(figsize = (100,100))
  wc.generate(all_words)
  plt.imshow(wc)
  plt.axis('off')
```

```python
from transformers import pipeline

# Load emotion classification pipeline
emotion_classifier = pipeline("text-classification", model="bhadresh-savani/bert-base-uncased-emotion", truncation=True)

def get_top_emotion(text):
    """Classify text into emotions and return the top predicted emotion."""
    text = text[:500]  # Truncate text to avoid token limit issues

    emotions = emotion_classifier(text)

    return max(emotions, key=lambda x: x["score"])["label"] if emotions else "unknown"
```

Due to limited CPU power and time, we randomly selected 100 reviews from the merged dataset. We used the following prompt to extract the main topics: “In the following customer review, pick out the main 3 topics. Return them in a numbered list format, with each one on a new line.”

```python
import json
import numpy as np
import random
from tqdm import tqdm  # For progress tracking
from transformers import pipeline  # Ensure this is imported

# Set random seed for reproducibility
random.seed(42)


# Define the prompt template
prompt_template = (
    "In the following customer review, pick out the main 3 topics. "
    "Return them in a numbered list format, with each one on a new line.\n\n"
    "Review: {}"
)

# Prepare reviews list from the DataFrame
bad_reviews_list_clean = bad_reviews_df["cleaned_review"].dropna().tolist()

# Randomly sample 100 reviews
sampled_reviews = random.sample(bad_reviews_list_clean, 100)

# List to store extracted topics
topics_llm = []

# Process each review
for review in tqdm(sampled_reviews, desc="Processing Reviews"):
    # Construct the full prompt
    prompt = prompt_template.format(review)

    # Get response from the model
    try:
        output = text_generator(prompt, max_length=1000)[0]["generated_text"]
        topic_list = output.strip().split("\n")  # Convert to list by splitting lines
    except Exception as e:
        topic_list = ["Error parsing response", str(e)]  # Handle unexpected errors

    # Store result
    topics_llm.append(topic_list)

# Convert list to NumPy array and save
np_topics_llm = np.array(topics_llm, dtype=object)  # Use dtype=object for variable-length lists
np.save('np_topics_phi.npy', np_topics_llm)

# Display a sample
print(np_topics_llm[:5])  # Show first 5 processed results
```

Create dictionary and corpora using Gensim

```python
# Create a dictionary representation of the documents.
from gensim import corpora
dictionary = corpora.Dictionary(tokenized_docs)


# Filter out words that occur fewer than 2 documents or more than 50% of the documents.
dictionary.filter_extremes(no_below=2, no_above=0.5)
```

Using Gensim, perform LDA on the tokenised data. Specify the number of topics = 10.

```python
# Create a BOW representation of the documents.
corpus = [dictionary.doc2bow(doc) for doc in tokenized_docs]

# Set parameters.
num_topics = 10
passes = 20
```

```python
# Create the LDA model.
from gensim.models.ldamodel import LdaModel
lda_model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=passes)
```
